```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# leer Indicators.csv
df = pd.read_csv('Indicators.csv')
df
```

```python
## funcion para graficar Year y Value a partir de los prarametros de country_name e indicator
def plot_country_indicator(country_name, indicator):
    # filtrar por country_name e indicator y plotear
    datos = df[(df['CountryName'] == country_name) & (df['IndicatorName'].str.startswith(indicator))][['Year', 'Value']]
    datos.plot(x='Year', y='Value', kind='line', title=country_name + ' ' + indicator, grid=True)

# probamos la funcion
plot_country_indicator('United States', 'CO2 emissions (metric tons per capita)')
plot_country_indicator('Guatemala', 'CO2 emissions (metric tons per capita)')

```

```python
## funcion para graficar 2 paises comparados en  Year y Value a partir de los prarametros de country_name e indicator
def plot_country_indicator(country_name1, country_name2, indicator):
    # filtrar por country_name e indicator y plotear
    datos1 = df[(df['CountryName'] == country_name1) & (df['IndicatorName'].str.startswith(indicator))][['Year', 'Value']]
    datos2 = df[(df['CountryName'] == country_name2) & (df['IndicatorName'].str.startswith(indicator))][['Year', 'Value']]
    
    plt.figure(figsize=(12,6))
    plt.plot(datos1['Year'], datos1['Value'], label=country_name1)
    plt.plot(datos2['Year'], datos2['Value'], label=country_name2)
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title(country_name1 + ' vs ' + country_name2 + ' ' + indicator)
    plt.legend()
    plt.show()


# probamos la funcion
plot_country_indicator('United States', 'China', 'CO2 emissions (metric tons per capita)')
```
