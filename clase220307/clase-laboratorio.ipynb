## 1. ¿Cuál es la relación POBLACION x LOTE? 

```python
df2=df[["nombre","municipio","poblacion", "n_lotes"]].sort_values(by=["poblacion"], ascending=False)
# asumimos que todos los poblados tienen al menos un lote (elimina error de division por cero)
df2[df2['n_lotes'] == 0] = 1
# calculamos la relación de poblacion por lote
df2["rel_pob_lote"] = df2["poblacion"] / df2["n_lotes"]
df2.sort_values(by=["rel_pob_lote"], ascending=False)
```

## 2. Por clasificación, ¿qué municipios cuentan con agua y  no tienen drenaje? 

```python
#Por clasificación, ¿Qué municipios cuentan con agua y no tienen drenaje?
#Convertir agua y drenajes a booleanos
df['n_agua'] = df['agua'] == 'Sí'
df['n_drenaje'] = df['drenaje'] == 'Sí'
#Definir nuevo dataframe con las columnas de mi interes
df_agdre = df[["municipio", "clasifica", "poblacion", "n_agua", "n_drenaje"]]
#Evaluando quienes tienen agua y no tienen drenajes
siag_nodre = df_agdre[(df_agdre["n_agua"]==True) & (df_agdre["n_drenaje"]==False)]
#Resultados
siag_nodre.groupby(['municipio'])[['clasifica']].count().sort_values(by=['clasifica'], ascending=[False])
```

## 3. ¿Cuál es el numero de personas por lote por cada una de las zonas?

```python
# Numero de personas por lote por cada una de las zonas? --> Donde tenemos mayor cantidad de gente  -> Dennis
# Paso 1: personas por lote por cada zona
df2 = df[['municipio', 'zona', 'n_lotes', 'poblacion']].groupby(['municipio', 'zona']).sum().sort_values(by=['poblacion'], ascending=[False])
df2['poblacion_por_lote'] = df2['poblacion'] / df2['n_lotes']
df2
```

## 4. ¿Dónde tenemos mayor cantidad de personas?

```python
# Paso 2: Dónde tenemos mayor cantidad de gente (continuación de la pregunta anterior)
df2.sort_values(by=['poblacion_por_lote'], ascending=[False])
```

```python
# Donde tenemos mayor cantidad de gente en general
df2.sort_values(by=['poblacion'], ascending=[False])
```

## 5. Cuáles son las características para tener agua, energía y drenaje?

```python
# marcar agua energia y drenaje como True o False
df['tiene_agua'] = df['agua'] == 'Sí'
df['tiene_energia'] = df['energia'] == 'Sí'
df['tiene_drenaje'] = df['drenaje'] == 'Sí'

# buscar los datos que cumplen con los tres servicios
df[df['tiene_agua'] & df['tiene_energia'] & df['tiene_drenaje']][['nombre', 'municipio', 'zona', 'poblacion', 'clasifica', 'tiene_agua', 'tiene_energia', 'tiene_drenaje']]
```

## 6. ¿Cuánta es la cantidad de población que vive por clasificación?

## 7. Según el año de fundación, ¿Cuál es el municipio que tiene mayor crecimiento de población?

```python
# dejar solo valores enteros en el año de fundación
df['fundacion'] = df.a_o_conf.str.replace('[a-z\sA-Z]', '', regex=True).astype(int)

# calcular la cantidad de años desde la fundacion hasta el día de hoy
df['años_desde_fundacion'] = datetime.now().year - df['fundacion']

# crecimiento anual de cada comunidad por años de fundacion
df['crecimiento_anual'] = df['poblacion'] / df['años_desde_fundacion']
df[['municipio', 'nombre', 'poblacion', 'años_desde_fundacion', 'crecimiento_anual']].sort_values(['crecimiento_anual'], ascending=False)
```

```python
# promedio de crecimiento por municipio
df[['municipio', 'años_desde_fundacion', 'crecimiento_anual']].groupby(['municipio']).mean().sort_values(['crecimiento_anual'], ascending=False)
```

## 8. ¿Cómo es la población por clasificación y sus servicios? 

```python
# contar poblados por clasificacion
total_por_clasificacion = df[['clasifica', 'municipio']].groupby(['clasifica']).count().sort_values(by=['municipio'], ascending=[False]) 
# renombrar municipio por total_clasificacion
total_por_clasificacion.rename(columns={'municipio': 'total_clasificacion'}, inplace=True)

# sumar por clasificacion
suma_de_servicios = df[['clasifica', 'poblacion', 'tiene_agua', 'tiene_energia', 'tiene_drenaje']].groupby(['clasifica']).sum()
poblados = suma_de_servicios.join(total_por_clasificacion)

# alcular porcentaje de servicios por poblacion
poblados['porcentaje_agua'] = poblados['tiene_agua'] / poblados['total_clasificacion'] 
poblados['porcentaje_energia'] = poblados['tiene_energia'] / poblados['total_clasificacion']
poblados['porcentaje_drenaje'] = poblados['tiene_drenaje'] / poblados['total_clasificacion']
poblados[['poblacion', 'porcentaje_agua', 'porcentaje_energia', 'porcentaje_drenaje']].sort_values(by=['porcentaje_agua'], ascending=[False])
```

## 13. ¿Qué % de la población tienen agua, energía y drenajes? 

```python
#¿Qué % de la población tiene agua, energía y drenajes
df_servp = df[["poblacion","agua","energia","drenaje"]]
#filtro valores "Sí"
df_servp = df_servp[df_servp["energia"]=="Sí"]
df_servp = df_servp[df_servp["agua"]=="Sí"]
df_servp = df_servp[df_servp["drenaje"]=="Sí"]

pob_general = sum(df["poblacion"])
pob_servicios = sum(df_servp["poblacion"])

porcentaje_servicios =(pob_servicios/pob_general)*100

print("El porcentaje de población que posee agua, energía y drenajes es del ", porcentaje_servicios)
```

## 14. ¿Qué zona es la mas descuidada?

```python
# sumar tiene_agua, tine_energia y tiene_drenaje
suma_servicios_por_zona = df[['zona', 'tiene_agua', 'tiene_energia', 'tiene_drenaje']].groupby(['zona']).sum().sort_values(by=['tiene_agua'], ascending=[False])
# sumar totalizado
suma_servicios_por_zona['servicios_totales'] = suma_servicios_por_zona['tiene_agua'] + suma_servicios_por_zona['tiene_energia'] + suma_servicios_por_zona['tiene_drenaje']
# ordenar y seleccionar la zona con menos servicios
suma_servicios_por_zona.reset_index(inplace=True)
suma_servicios_por_zona.sort_values(by=['servicios_totales'], ascending=True).iloc[0]
```

## 15. ¿Qué municipio es el peor cuidado? 

```python
# sumar tiene_agua, tine_energia y tiene_drenaje
servicios_por_municipio = df[['municipio', 'tiene_agua', 'tiene_energia', 'tiene_drenaje']].groupby(['municipio']).sum().sort_values(by=['tiene_agua'], ascending=[False])
# sumar totalizado
servicios_por_municipio['servicios_totales'] = servicios_por_municipio['tiene_agua'] + servicios_por_municipio['tiene_energia'] + servicios_por_municipio['tiene_drenaje']
# ordenar y seleccionar la zona con menos servicios
servicios_por_municipio.reset_index(inplace=True)
servicios_por_municipio.sort_values(by=['servicios_totales'], ascending=True).iloc[0]
```

## 17. ¿Qué porcentaje de la población cuenta con energía por municipio? 

```python
# calcular poblacion total
poblacion_total = df[['municipio', 'poblacion']].groupby('municipio').sum()

# calcular poblacion con energia
poblacion_con_energia = df[df['tiene_energia']][['municipio','poblacion']].groupby('municipio').sum()
# renombrar poblacion por poblacion_con_energia
poblacion_con_energia.rename(columns={'poblacion': 'poblacion_con_energia'}, inplace=True)

# join y calculo del porcentaje
poblacion_porcentaje = poblacion_con_energia.join(poblacion_total)
poblacion_porcentaje['porcentaje'] = poblacion_porcentaje['poblacion_con_energia'] / poblacion_porcentaje['poblacion'] * 100
poblacion_porcentaje.sort_values(by=['porcentaje'], ascending=False)
```

## 18. ¿Cuál es la proporción agua y drenaje de cada municipio? 

```python
# calcular la proporcion de poblados que tienen agua vs poblados que tienen drenaje
agua_drenaje = df[['municipio', 'tiene_agua', 'tiene_drenaje']].groupby('municipio').sum()
agua_drenaje['proporcion'] = agua_drenaje['tiene_agua'] / agua_drenaje['tiene_drenaje']
agua_drenaje
```

## 19. ¿Qué % de la población vive en cada uno de los municipios? 

```python
# sumar poblacion agrupada por poblado
poblacion_por_municipio = df[['municipio', 'poblacion']].groupby('municipio').sum()
# calcular la proporcion de poblacion
poblacion_por_municipio['proporcion'] = poblacion_por_municipio['poblacion'] / poblacion_por_municipio['poblacion'].sum() * 100
poblacion_por_municipio
```
