# Pregunta 1: ¿Qué mes es el que tiene la mayor demanda?

```python
# cnt es el total de ventas por hora, sumamos cnt por cada mes
suma_por_mes = df[['cnt', 'mnth']].groupby('mnth').sum()
print(suma_por_mes.sort_values('cnt', ascending=False).head(6))
suma_por_mes.plot(figsize=(20,15))
```

## Respuesta: Agosto

Agosto es el mes con mayor demanda, pero cabe resaltar que la demanda es establemente alta entre mayo y octubre

# Pregunta 2: ¿Qué rango de hora es la de mayor demanda?

```python
# promedio por hora
avg_hr = df[['cnt', 'hr']].groupby('hr').mean()
print(avg_hr.sort_values('cnt', ascending=False).head())
avg_hr.plot(figsize=(20,15))
```

## Respuesta:  5pm
La hora pico es 5pm, el rango mas alto es entre 5 y 6pm, con un pico también a las 8am. Presumiblemente a la hora de ingreso y salida del horario laboral.

# Pregunta 3:  ¿Qué temporada es la mas alta?

```python
# sumamos por season
cnt_season = df[['cnt', 'season']].groupby('season').sum()
print(cnt_season.sort_values('cnt', ascending=False).head())
cnt_season.plot(figsize=(20,15), kind='bar')
```

## Respuesta:  La temporada 3

Si suponemos temporadas en el orden primavera, verano, otoño, invierno, la temporada mas alta es **otoño**, verano es la segunda.

Esto tiene sentido si comparamos con el resultado por meses donde tenemos la mayor venta entre mayo - octubre


# Pregunta 4: ¿A qué temperatura baja la demanda?

```python
# tmp puede tomarse como categorico? --> 50, valores, lo usaremos como categoria
print(len(df['temp'].unique()))

# suma de cnt por temperatura
prom_por_temp = df[['cnt', 'temp']].groupby('temp').sum()
print(prom_por_temp.sort_values('cnt').head(10))
prom_por_temp.plot(figsize=(20,15))
```

## Respuesta: La demanda baja en ambos extremos de temperatura

Podemos observar que ambos las temperatura más altas y bajas producen un decenso considerable de la demanda.

# Pregunta 5: ¿A qué humedad baja la demanda?

```python
# promedio de cnt por humedad
cnt_hum = df[['cnt', 'hum']].groupby('hum').sum()
print(cnt_hum.sort_values('cnt').head(10))
cnt_hum.plot(figsize=(20,15))
```

## Respuesta:
La demanda es muy baja cuando la humedad es muy baja (<0.2), pero también decrece en relación directa con el aumento de humedad a partir de 0.2

# Pregunta 6: ¿Qué condiciones serian ideales para nuestra demanda?

```python
# pivot de cnt por temperatura y humedad, sumando ventas
pivot_cnt_temp_hum = df[['cnt', 'temp', 'hum']].pivot_table(index='temp', columns='hum', values='cnt', aggfunc='sum')
print(pivot_cnt_temp_hum.max().max())

# el rango llega hasta > 22000, seleccionemos el rango donde vendemos al menos la mitad, 11000
plt.figure(figsize=(20,15))
sns.heatmap(pivot_cnt_temp_hum[pivot_cnt_temp_hum > 11000], vmin=0)
```

## Respuesta:

Dadas las observaciones, nuestra combinación ideal sería:

* Meses entre mayo y octubre, especialmente en otoño
* Temperatura entre 0.5 y 0.85
* Humedad entre 0.40 y 0.89
* Horario al rededor de las 8am y 5pm.
