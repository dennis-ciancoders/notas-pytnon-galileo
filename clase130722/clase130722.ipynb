```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.png')
```

```python
# corregir el formato de color de la imagen BGR a RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
img = cv2.imread('image.png')

blue = img[:, :, 0]
green = img[:, :, 1]
red = img[:, :, 2]

grayscale = 0.299*blue + 0.587*green + 0.114*red
grayscale.astype(np.uint8)
plt.imshow(grayscale, cmap='gray')
```

```python
alto = img.shape[0]
ancho = img.shape[1]

lienzo = np.zeros((alto, ancho, 1), dtype=np.uint8)

for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]

        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]

        lienzo[i,j] = int(0.299*blue + 0.587*green + 0.114*red)

plt.imshow(lienzo)
```

```python
img = cv2.imread('image.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_rgb, cmap='gray')
```
