```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import plotly.express as px
```

```python
# leer players.csv
players = pd.read_csv('players.csv', parse_dates=['date_of_birth'])

# leer clubs.csv
clubs = pd.read_csv('clubs.csv')

# leer competitions.csv
competitions = pd.read_csv('competitions.csv')

# leer games.csv
games = pd.read_csv('games.csv')
```

```python
# edad en base a la fecha de nacimiento
players['edad'] = players['date_of_birth'].apply(lambda x: datetime.now().year - x.year)
```

```python
players[['market_value_in_gbp', 'height_in_cm',  'edad']]

# correlation matrix
corr = players.corr()
display(corr)

plt.figure(figsize=(15,15))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

players.dropna(inplace=True)
px.histogram(players, x='edad', marginal='box', color='position')
```
